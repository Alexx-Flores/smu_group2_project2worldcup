```python
cols = ['name', 'position', 'squad', 'age', 'club', 'birth_year',
       'matches_played', 'starts', 'minutes_played']
df_players_for_main = dfplayers.loc[:,cols]
df_keepers_for_main = dfkeepers.loc[:,cols]
```

```python
dftotalplayers_main_table = pd.concat([df_players_for_main, df_keepers_for_main]).reset_index(drop=True)
dftotalplayers_main_table.head()
```

```python
for i in range(len(dftotalplayers_main_table)):
    dftotalplayers_main_table.loc[i, ["squad"]] = dftotalplayers_main_table.squad.values[i][3:]
```

```python
for j in range(len(dftotalplayers_main_table)):
    if dftotalplayers_main_table.club.values[j] == dftotalplayers_main_table.club.values[j]:
        a = dftotalplayers_main_table.club.values[j].find(' ')
        dftotalplayers_main_table.loc[j, ["club"]] = dftotalplayers_main_table.club.values[j][a+1:]
```

```python
dftotalplayers_main_table["position_2"] = None
```

```python
for k in range(len(dftotalplayers_main_table)):
    pos_list = dftotalplayers_main_table.position.values[k].split(',')
    if len(pos_list) == 2:
        dftotalplayers_main_table.loc[k, "position"] = pos_list[0]
        dftotalplayers_main_table.loc[k, "position_2"] = pos_list[1]
    
```

```python
dftotalplayers_main_table["id"] = np.arange(1,len(dftotalplayers_main_table)+1)
```

```python
squads_dict = {}
for x in range(len(df_squads)):
    squads_dict[df_squads.squads.tolist()[x]] = df_squads.id.tolist()[x]
```

```python
positions_dict = {}
for x in range(len(df_positions)):
    positions_dict[df_positions.positions.tolist()[x]] = df_positions.id.tolist()[x]
```

```python
clubs_dict = {}
for x in range(len(df_clubs)):
    clubs_dict[df_clubs.clubs.tolist()[x]] = df_clubs.id.tolist()[x]
```

```python
for i in range(len(dftp)):
    dftp.iloc[i, 4] = squads_dict[dftp.iloc[i, 4]]
```

```python
for i in range(len(dftp)):
    dftp.iloc[i, 2] = positions_dict[dftp.iloc[i, 2]]
```

```python
for i in range(len(dftp)):
    if (dftp.iloc[i, 3] != None):
        dftp.iloc[i, 3] = positions_dict[dftp.iloc[i, 3]]
```

```python
for i in range(len(dftp)):
    dftp.iloc[i, 6] = clubs_dict[dftp.iloc[i, 6]]
```

```python
dftp = dftp.rename(columns = {"position": "position_id", "position_2": "position_2_id", "club": "club_id",
                      "squad": "squad_id"})
```

```python
player_name_dict = {}
for x in range(len(dftp)):
    player_name_dict[dftp.name.tolist()[x]] = dftp.id.tolist()[x]
```

```python
for i in range(len(df_players_stats)):
    df_players_stats.iloc[i, 1] = player_name_dict[df_players_stats.iloc[i, 1]]
```

```python
df_players_stats = df_players_stats.rename(columns = {"name": "player_id"})
```

```python
for i in range(len(df_keepers_stats)):
    df_keepers_stats.iloc[i, 1] = player_name_dict[df_keepers_stats.iloc[i, 1]]
```

```python
df_keepers_stats = df_keepers_stats.rename(columns = {"name": "player_id"})
```
